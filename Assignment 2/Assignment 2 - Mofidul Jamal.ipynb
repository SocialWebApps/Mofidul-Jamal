###Getting bit.ly data
The following segment does the following:
* Loads the 5 bit.ly data files
* Parses them and stores them into a single list
* Loads the list into a pandas DataFrame object
* Filters the DataFrame object for valid longitude/lattitude pairs

```python
import pandas as pd
import numpy as np
import json
from pandas import DataFrame, Series
import csv

files = ["datafiles\\usagov_bitly_data2013-05-17-1368817803", 
         "datafiles\\usagov_bitly_data2013-05-17-1368821404", 
         "datafiles\\usagov_bitly_data2013-05-17-1368825005", 
         "datafiles\\usagov_bitly_data2013-05-17-1368828605",
         "datafiles\\usagov_bitly_data2013-05-17-1368832207"]
datarecords = []

#load all 5 files records into a single object
for filename in files:
    records = [json.loads(line) for line in open(filename)]
    datarecords = datarecords + records
#now load it up into a pandas dataframe
df = DataFrame(records)
#we are going to filter the data so that only valid records with long/lat info are present
#we filter it this way because the data record is either NaN or a list with 2 elements for longitude and lattitude
locations = filter(lambda record: type(record) == list, df['ll'])
```

### Plotting the points using folium, but first helper funcs
We are going to plot the points using the folium library. In order to do that we have to embed a little html into python using the following helper functions which are lifted directly off the folium docs.

```python
from IPython.display import HTML
import folium

def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))

def embed_map(map, path="m213map.html"):
    """
    Embeds a linked iframe to the map into the IPython notebook.
    
    Note: this method will not capture the source of the map into the notebook.
    This method should work for all maps (as long as they use relative urls).
    """
    map.create_map(path=path)
    return HTML('<iframe src="files/{path}" style="width: 100%; height: 510px; border: none"></iframe>'.format(path=path))
```

### Plot map using foliium, centered on Eastern US
We are going to provide the folium library with all our lat/long pairs and set them to be clustered automatically. This will provide an interactive map that clusters points and provides counts automatically and is fully interactive at various zoom levels. Try it out!

```python
map = folium.Map(width=800,height=500,location=[44, -73], zoom_start=4)
for record in locations:
    map.simple_marker(record, popup = str(record), clustered_marker=True)
#use the helper func to display the map ipython notebook
inline_map(map)
```
