###Connecting to LinkedIn, saving connections data to disk
This file will do the following:
* Connect to LinkedIn using api keys
* Use the LinkedIn Connections API to request the member id's of your connections
* Use the member ID's to query LinkedIn for the names of the titles and companies for each member in your connections
* Write out the member ID's to a file named _linkedin_connections_data.json_ located in the data folder (json formatted)

I've elected to only keep the member ID's as this is really the only piece of information we need to query for job titles. I only save the member ID, job title and company name to the file.

```python
from linkedin import linkedin # pip install python-linkedin
import json
from prettytable import PrettyTable

# Define CONSUMER_KEY, CONSUMER_SECRET,  
# USER_TOKEN, and USER_SECRET from the credentials 
# provided in your LinkedIn application
CONSUMER_KEY = ''
CONSUMER_SECRET = ''
USER_TOKEN = ''
USER_SECRET = ''

RETURN_URL = '' # Not required for developer authentication

# Instantiate the developer authentication class

auth = linkedin.LinkedInDeveloperAuthentication(CONSUMER_KEY, CONSUMER_SECRET, 
                                USER_TOKEN, USER_SECRET, 
                                RETURN_URL, 
                                permissions=linkedin.PERMISSIONS.enums.values())

# Pass it in to the app...
app = linkedin.LinkedInApplication(auth)
#get connections (get member ID's only, save to file)
my_connections = app.get_connections(selectors=['id'])
#print it out just to verify it exists
print json.dumps(my_connections, indent=1)
```

### Using member id's to request job titles/companies for those members of your connections
The following block will:
* Enumerate the member ID's and request job titles/companies for each member of your connections that was obtained in the previo0us step



```python
import time
#will hold all the data
member_data = []
for member_record in my_connections['values']:
    id = member_record['id'] 
    if id != 'private': #some of them are private, get a 404 error if you try to look up private info
        member_positions = app.get_profile(member_id=id, 
                                       selectors=['id,positions:(title,company:(name))'])
        member_data.append(member_positions)
```

```python
#print out the data just to view it
print json.dumps(member_data, indent=1)
```

###Saving data to file, opening it up and printing it out
The following block of code will:
* save all the data gathered in the previous steps to a file 
* open up the file it just wrote
* parse the file and use pretty tables to dump out the gathered data for your viewing pleasure

```python
connections_data = 'data\\linkedin_connections_data.json'
f = open(connections_data, 'w')
f.write(json.dumps(member_data, indent=1))
f.close()

#now open up the file we just wrote to print out the data and verify everything was written out nicely
with open('data\\linkedin_connections_data.json') as data_file:
    json_data = json.load(data_file)
    pt = PrettyTable(field_names=['Id', 'Company', 'Title'])
    titles = []
    for entry in json_data:
        #print entry
        if 'values' in entry['positions']:
            for position in entry['positions']['values']:
                if 'name' in position['company']:
                    pt.add_row([entry['id'],position['company']['name'], position['title']])
                    titles.append({'title': position['title'], 'id': entry['id'], 'company': position['company']['name']})

#print it out
print pt
```
