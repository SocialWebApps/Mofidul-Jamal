#Querying Twitter
The following block of code will:
* Use the twitter api to search for 1.usa.gov URLS
* Parse through 5000 records
* Verify if the URL has geographical information
* If the URL has geo info, save the URL and coordinates

```python
from TwitterAPI import TwitterAPI, TwitterRestPager

#Please fill in with your own API keys if you use this code
SEARCH_TERM = '1.usa.gov'
ACCESS_TOKEN_KEY = 'YOUR ACCESS KEY'
ACCESS_TOKEN_SECRET = 'YOUR SECRET'
CONSUMER_KEY = 'YOUR CONSUMER KEY'
CONSUMER_SECRET = 'YOUR CONSUMER SECRET'

api = TwitterAPI(CONSUMER_KEY,
                 CONSUMER_SECRET,
                 ACCESS_TOKEN_KEY,
                 ACCESS_TOKEN_SECRET)
r = TwitterRestPager(api, 'search/tweets', {'q': SEARCH_TERM , 'include_entities':True, 'count':100})
print r.get_iterator()
total_requests = 0
coordinates_found = 0
items_found = []
for item in r.get_iterator():
    item_dict = {}
    if total_requests == 100*50: #100 * 50 = 5000 records to parse
        break;
    total_requests += 1
    if 'message' in item and item['code'] == 88:
        print 'SUSPEND, RATE LIMIT EXCEEDED: %s\n' % item['message']
        break
    if 'entities' in item and 'geo' in item and item['geo'] != None:
        if 'urls' in item['entities']:
            for url in item['entities']['urls']:
                if "1.usa.gov" in url['expanded_url']:
                    print url
                    print item['geo']
                    if 'coordinates' in item['geo']:
                        #we have to transpose the data because twitter gives lat/long
                        #and folium expects long/lat
                        item_dict['ll'] = [item['geo']['coordinates'][1], item['geo']['coordinates'][0]]
                        item_dict['url'] = url['display_url']
                        items_found.append(item_dict)
                        coordinates_found +=1
print "\nItemsParsed %d, coords found %d\n" % (total_requests, coordinates_found)
print "You may continue"
```

### Plotting the points using folium, but first helper funcs
We are going to plot the points using the folium library. In order to do that we have to embed a little html into python using the following helper functions which are lifted directly off the folium docs.

```python
from IPython.display import HTML
import folium

def inline_map(map):
    """
    Embeds the HTML source of the map directly into the IPython notebook.
    
    This method will not work if the map depends on any files (json data). Also this uses
    the HTML5 srcdoc attribute, which may not be supported in all browsers.
    """
    map._build_map()
    return HTML('<iframe srcdoc="{srcdoc}" style="width: 100%; height: 510px; border: none"></iframe>'.format(srcdoc=map.HTML.replace('"', '&quot;')))

def embed_map(map, path="m213map.html"):
    """
    Embeds a linked iframe to the map into the IPython notebook.
    
    Note: this method will not capture the source of the map into the notebook.
    This method should work for all maps (as long as they use relative urls).
    """
    map.create_map(path=path)
    return HTML('<iframe src="files/{path}" style="width: 100%; height: 510px; border: none"></iframe>'.format(path=path))
```

### Plot map using foliium, centered on Eastern US
We are going to provide the folium library with all our lat/long pairs

```python
map = folium.Map(width=800,height=500,location=[44, -73], zoom_start=4)
for record_dict in items_found:
    if record_dict['ll'] != [0.0,0.0]:
        map.simple_marker(record_dict['ll'], popup = str(record_dict['url']))
#use the helper func to display the map ipython notebook
inline_map(map)
```
